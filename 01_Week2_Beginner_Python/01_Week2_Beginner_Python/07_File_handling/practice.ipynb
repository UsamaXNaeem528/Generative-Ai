File Handling with .txt files

```python
list = []
with open(r'file.txt','r') as file:
    for line in file:
        list.append(line.strip())
    print(list)
```

```python
lines_to_append = [
    "First line\n",
    "Second line\n",
    "Third line\n",
]

with open(r'file.txt', 'a') as f:
    for line in lines_to_append:
        f.write(line)
```

```python
"""
Question:

You are given a CSV file scores.csv where each line contains a player's name followed by their score,
separated by a comma (e.g., Alice,45). Write a Python program to read this file and do the following:

Store the scores of each player in a dictionary,
where the key is the player's name and the value is a list of their scores.

For each player, calculate and display:
The minimum score
The maximum score
The average score

Sample output format:
Alice ===> Min:30, Max:90, Avg:60.0
Bob ===> Min:40, Max:70, Avg:55.0
"""
```

```python
player_scores = {}
with open(r'scores.csv', 'r') as f:
    data=csv.reader(f)
    for player in data:
        name = player[0]
        score = int(player[1])
        # print(name, score)
        if name in player_scores:
            player_scores[name].append(score)
        else:
            player_scores[name] = [score]

print(player_scores)        
```

```python
for name, score in player_scores.items():
    min_score = min(score)
    max_score = max(score)
    avg_score = round(sum(score)/len(score),2)
    print(f'{name} ==> Min:{min_score}, Max:{max_score}, Avg:{avg_score}')
```
