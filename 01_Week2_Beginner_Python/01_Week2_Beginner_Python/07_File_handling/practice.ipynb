```python
list = []
with open(r'file.txt','r') as file:
    for line in file:
        list.append(line.strip())
    print(list)
```

```python
with open(r'file.txt','r') as file:
    content = file.read()
print(content)
```

```python
with open(r'file.txt', 'a') as f:
    f.write('Hello')
```

```python
lines_to_append = [
    "First line\n",
    "Second line\n",
    "Third line\n",
]

with open(r'file.txt', 'a') as f:
    for line in lines_to_append:
        f.write(line)
```
