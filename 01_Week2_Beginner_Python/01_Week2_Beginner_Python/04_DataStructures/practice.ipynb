```python
#Question1
'''
We have following information on countries and their population (population is in crores),

Country	Population
China	143
India	136
USA	    32
Pakistan	21

i. Using above create a dictionary of countries and its population
ii. Write a program that asks user for three type of inputs,
    a. print: if user enter print then it should print all countries with their population in this format,
    china==>143
    india==>136
    usa==>32
    pakistan==>21
    
    b. add: if user input add then it should fursther ask for a country name to add. If country already exist in our dataset then it should print that it exist and do nothing. If it doesn't then it asks for population and add that new country/population in our dictionary and print it
    c. remove: when user inputs remove it should ask for a country to remove. If country exist in our dictionary then remove it and print new dictionary using format shown above in (a). Else print that country doesn't exist!
    d. query: on this again ask user for which country he or she wants to query. When user inputs that country it will print population of that country.
'''
```

```python
population = {
    'china' : 143,
    'india' : 136,
    'usa' : 32,
    'pakistan' : 21
}

def add_country():
    country_to_add = input(f'Enter country name : ').lower()
    if country_to_add in population:
        print(f'Country already exist! You cannot add') 
        return
    else:
        country_population = int(input(f'Enter Poplulation of country in digits : '))
        population[country_to_add] = country_population
        print(f'Country Successfully Added!')
        for key, value in population.items():
            print(f'{key} ==> {value}')

    
def del_country():
    country_to_remove =  input(f'Enter country name to remove : ').lower()
    for country in population.keys():
        if country_to_remove == country:
            population.pop(country_to_remove)
            print('Country Successfull Deleted')
            return
    else:
        print(f'Country dont exist in populatation dict.')

def user_query():
    query_country = input(f'Enter country name you want to query : ').lower()
    if query_country in population:
        print(f'{query_country} : {population[query_country]}')
        return
    else:
        print(f'{query_country} country does\'t exits')

def print_all():
    for key, value in population.items():
            print(f'{key} ==> {value}')

def main():
    while True:
        print('Choose option print / add / remove / query')
        choice = input(f'Enter your option : ').lower()
        if choice == 'add':
            user_query()
        elif choice == 'remove':
            del_country()
        elif choice == 'query':
            user_query()
        else: print(f'Enter the correct option')
```
