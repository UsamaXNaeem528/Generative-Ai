```python
'''
🔥 1. Nested Exceptions (File + Data Processing)

You are given a file that contains numbers (one per line). Some lines may contain invalid data.

Task:

Read the file
Convert each line to an integer
Skip invalid lines
Print sum of valid numbers
If file is missing, print a custom message

Extra challenge: log how many invalid lines were found.
'''
```

```python
total = 0
invalid_count = 0
try:
    with open(r'C:\Users\Admin\Downloads\numbers.txt', 'r') as file:
        for line in file:
           line = line.strip()

           try:
            num = int(line)
            total += num
           except ValueError:
            invalid_count += 1
    
    print("Sum of valid number", total)
    print("Sum of invalid count", invalid_count)

        
except FileNotFoundError:
    print("File not found in directory")
```
