'''Task is that take 10 students marks from user and print such students marks having marks is greater than
or equal to 80 marks in list form '''

```python
marks_lst = []

for i in range(10):
    i += 1
    marks = int(input(f'Enter {i} Student Marks : '))
    marks_lst.append(marks)
```

```python
for marks in marks_lst:
    if marks >= 80:
        print(marks)
```
